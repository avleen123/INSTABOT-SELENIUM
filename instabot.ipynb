## Q5. Like/Unlike posts

### Q5.1 Liking the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it.

```python
def like():
    search1=driver.find_element_by_class_name('XTCLo.x3qfX')
    # explicitly waiting until the search box is enabled
    search = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "XTCLo")))
    search.send_keys('dilsefoodie')
    # Explicitly waiting until the search result loads
    WebDriverWait(driver,10).until(EC.presence_of_element_located((By.CLASS_NAME,"fuqBx")))
    driver.find_element_by_class_name('Ap253').click()
    #Clicking on the first post and liking it.
    first_post = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME , '_9AhH0')))
    first_post.click()
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH , '//section[contains(@class , "ltpMr")]/span[1]/button'))).click()
    count=1
    for i in range(30):
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME,'coreSpriteRightPaginationArrow'))).click()
        #WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH,'//section[contains(@class , "ltpMr")]/span[1]/button'))).click()
        like_btn = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH,'//*[contains(@aria-label,"ike")]'))) 
        if like_btn.get_attribute("aria-label") == 'Like':
            like_btn.click() 
        else:
            print(count,'already liked')
            count+=1
#calling fn to like top 30 posts of the 'dilsefoodie'.Print message if u have already liked it.
like()
```

## Q6. Extract list of followers

### Q6.1 Extract the usernames of the first 500 followers of ‘foodtalkindia’ and ‘sodelhi’.

```python
def get_followers_username(handle_name):
    driver.get('https://instagram.com/'+handle_name)

    # wait till page is loaded
    wait = WebDriverWait(driver,20) #all imports done on top . No need to import again
    followers = wait.until(EC.presence_of_element_located( (By.PARTIAL_LINK_TEXT, 'followers') ))
    driver.execute_script('arguments[0].click()',followers)        

    time.sleep(10) # wait to get list of followers displayed

    while True:
        # locate any element on pop_up page
        
        WebDriverWait(driver,20).until(EC.presence_of_element_located( (By.XPATH, '//a[contains(@class,"FPmhX notranslate")]') ))
        obj_created_for_pop_up_scroll = driver.find_element_by_xpath('//a[contains(@class,"FPmhX notranslate")]')
        followers_list = driver.find_elements_by_class_name('FPmhX.notranslate._0imsa')
        if len(followers_list) >= 500:
            break
        obj_created_for_pop_up_scroll.send_keys(Keys.END)
        time.sleep(2) # wait to load new followers

    for i in range(500):
        print(str(i+1)+") ",followers_list[i].get_attribute('innerHTML'))
        
```

```python
# Get usernames of followers of foodtalkindia
get_followers_username('foodtalkindia')
```

## Q7.Check the story of ‘coding.ninjas’. Consider the following Scenarios and print error messages accordingly 

### Q7.1 If You have already seen the story.
### Q7.2 Or The user has no story.
### Q7.3 Or View the story if not yet seen.

```python
def check_story(handle_name):
    # locating and enabling the search by clicking on it
    search_enable = driver.find_element_by_class_name('TqC_a')
    search_enable.click()
    # explicitly waiting until the search box is enabled
    search_box = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "XTCLo")))
    # writing "coding.ninjas" in search box which is passed as handle_name
    search_box.send_keys(handle_name)
    # Explicitly waiting until the search result loads
    WebDriverWait(driver,10).until(EC.presence_of_element_located((By.CLASS_NAME,"fuqBx")))
    # Openeing the profile of coding.ninjas by clicking on the search result
    for i in driver.find_elements_by_xpath('//span[@class="Ap253"]'):
        if i.get_attribute("innerHTML")==handle_name:
            i.click()
            break
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "_6VtSN")))
    time.sleep(2)
    try:
        story = driver.find_element_by_xpath("//div[@class='RR-M- h5uC0']")
        story_exists = True
    except:
        story_exists = False
    
    if story_exists:
        if driver.find_element_by_xpath("//div[@class='RR-M- h5uC0']/canvas").value_of_css_property("top")=="-8px":
            print("Story Already seen")
        else:
            story.click()
            print("Seeing story now")
    else:
        print("No story available")
#fn to check story of coding ninjas
check_story("coding.ninjas")
```
